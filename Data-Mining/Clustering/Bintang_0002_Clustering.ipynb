```python
# Import Library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Import Dataset
df = pd.read_csv('data_penjualan_pelanggan.csv')

# Menampilkan 5 data teratas
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Diskon_Diterima', y='Jumlah_Produk_Dibeli', s=100)
plt.title('Segmentasi Pelanggan berdasarkan Diskon Diterima dan Jumlah Produk Dibeli')
plt.xlabel('Diskon Diterima')
plt.ylabel('Jumlah Produk Dibeli')
plt.show()
```

```python
# Pilih Fitur yang Relevan
X = df[['Diskon_Diterima', 'Jumlah_Produk_Dibeli']]

# Latih model KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Diskon_Diterima', y='Jumlah_Produk_Dibeli', 
                hue='Cluster', palette='viridis', s=100)

# Centroid
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroid')

plt.title('Segmentasi Pelanggan dengan KMeans')
plt.xlabel('Diskon Diterima')
plt.ylabel('Jumlah Produk Dibeli')
plt.legend()
plt.show()
```
