```python
# Import Library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Import Dataset
df = pd.read_csv('data_penjualan_pelanggan.csv')

# Menampilkan 5 data teratas
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Total_Pembelian_Juta', y='Frekuensi_Kunjungan', s=100)
plt.title('Segmentasi Pelanggan berdasarkan Total Pembelian dan Frekuensi Pembelian')
plt.xlabel('Total Pembelian (juta)')
plt.ylabel('Frekuensi Kunjungan(perbulan)')
plt.show()
```

```python
# Pilih Fitur yang Relevan
X = df[['Total_Pembelian_Juta', 'Frekuensi_Kunjungan']]

# Latih model KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Total_Pembelian_Juta', y='Frekuensi_Kunjungan', 
                hue='Cluster', palette='viridis', s=100)

# Centroid
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroid')

plt.title('Segmentasi Pelanggan dengan KMeans')
plt.xlabel('Total Pembelian (juta)')
plt.ylabel('Frekuensi Kunjungan(perbulan)')
plt.legend()
plt.show()
```
